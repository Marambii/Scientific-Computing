```python
data.fillna(data['Sum of Life Expectancy  (both sexes)'].mean(), inplace=True)

data
```

```python
Q1 = data['Sum of Males  Life Expectancy'].quantile(0.25)
Q3 = data['Sum of Males  Life Expectancy'].quantile(0.75)

Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound, upper_bound
```

```python
data[data['Sum of Males  Life Expectancy'].between(lower_bound, upper_bound)]
data_no_outliers = data[~((data['Sum of Males  Life Expectancy'] < lower_bound) | (data['Sum of Males  Life Expectancy'] > upper_bound))]
data_no_outliers
```

```python
mean = data['Sum of Life Expectancy  (both sexes)'].mean()
median = data['Sum of Life Expectancy  (both sexes)'].median()
mode = data['Sum of Life Expectancy  (both sexes)'].mode()

mean, median, mode
```
